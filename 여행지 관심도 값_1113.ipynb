```python
import pandas as pd
import os

# 파일 경로 설정
folder_path = 'SepData'
file_names = ['Data_20230904.csv', 'Data_20230911.csv', 'Data_20230918.csv', 'Data_20230925.csv']

# 각 파일을 읽어와서 리스트에 저장
dfs = [pd.read_csv(os.path.join(folder_path, file_name)) for file_name in file_names]

# 리스트에 저장된 데이터프레임들을 하나로 합치기
df_concat = pd.concat(dfs, ignore_index=True)

# 합쳐진 데이터프레임 확인
print(df_concat.head())
```

```python
# xlsx 파일로 저장
df_concat.to_excel('1113.xlsx', index=False)
```

```python
# 컬럼명 변경
df_concat.rename(columns={'SEXDSTN_FLAG_CD': '성별', 'AGRDE_FLAG_NM': '연령대', 'ANSWRR_OC_AREA_NM': '거주지','EXAMIN_BEGIN_DE'
 : '조사일자'}, inplace=True)

# '성별' 컬럼의 값 변경
df_concat['성별'] = df_concat['성별'].replace({'F': '여자', 'M': '남자'})

# 결과 확인
print(df_concat.head())
df_concat.to_excel('1113.xlsx', index=False)
```

```python
import csv
# 컬럼명 변경
df_concat.rename(columns={
    'SU_TOUR_INTRST_VALUE': '서울',
    'GG_IC_TOUR_INTRST_VALUE': '경기',
    'BS_TOUR_INTRST_VALUE': '부산',
    'CHN_DJ_TOUR_INTRST_VALUE': '충남',
    'CB_TOUR_INTRST_VALUE': '충북',
    'GN_US_TOUR_INTRST_VALUE': '경남',
    'GB_DG_TOUR_INTRST_VALUE': '경북',
    'JN_GJ_TOUR_INTRST_VALUE': '전남',
    'JNBK_TOUR_INTRST_VALUE': '전북',
    'GW_TOUR_INTRST_VALUE': '강원',
    'JJ_TOUR_INTRST_VALUE': '제주도'
}, inplace=True)

# 데이터프레임을 리스트 형태로 변환
data_list = df_concat.values.tolist()

# csv 파일로 저장
with open('1113_2.csv', 'w', newline='', encoding='utf-8-sig') as f:
    writer = csv.writer(f)
    writer.writerows(data_list)

# xlsx 파일로 저장
df_concat.to_excel('1113_2.xlsx', index=False)
```

```python
# 'A', 'F' 컬럼 제거
df_concat.drop(columns=['RESPOND_ID', 'HSHLD_INCOME_DGREE_NM'], inplace=True)

# 특정 값 변경
value_map = {
    '많이 적어졌다': '1',
    '약간 적어졌다': '2',
    '예전과 비슷하다': '3',
    '약간 커졌다': '4',
    '많이 커졌다': '5'
}
df_concat.replace(value_map, inplace=True)

# 데이터프레임을 리스트 형태로 변환
data_list = df_concat.values.tolist()

# csv 파일로 저장
with open('1113_3.csv', 'w', newline='', encoding='utf-8-sig') as f:
    writer = csv.writer(f)
    writer.writerows(data_list)

# xlsx 파일로 저장
df_concat.to_excel('1113_3.xlsx', index=False)
df_concat.to_excel('1113_3.csv', index=False)
```

```python
# 성별, 연령대, 거주지별로 그룹화
grouped = df_concat.groupby(['성별', '연령대', '거주지'])

# 각 그룹에 대해 오름차순 또는 가나다순으로 정렬
sorted_df = grouped.apply(lambda g: g.sort_values(by=list(g.select_dtypes(include=[np.number]).columns), ascending=True) if g.select_dtypes(include=[np.number]).shape[1] > 0 else g.sort_values(by=list(g.select_dtypes(include=[object]).columns), ascending=True))

# 데이터프레임을 리스트 형태로 변환
data_list = sorted_df.values.tolist()

# 출력
for data in data_list:
    print(data)
df_concat.to_excel('1113_6.xlsx', index=False)
# csv 파일로 저장
with open('1113_6.csv', 'w', newline='', encoding='utf-8-sig') as f:
    writer = csv.writer(f)
    writer.writerows(data_list)
```

```python
# 각 그룹에 대해 '관심도'라는 이름의 컬럼으로 평균값을 계산
interest_df = grouped.mean().reset_index()
interest_df['관심도'] = interest_df.iloc[:,3:].mean(axis=1)

# '관심도' 컬럼만 유지
interest_df = interest_df[['성별', '연령대', '거주지', '관심도']]

# csv 파일로 저장
interest_df.to_csv('1113_5.csv', index=False, encoding='utf-8-sig')

# xlsx 파일로 저장
interest_df.to_excel('1113_5.xlsx', index=False)

# 데이터프레임 확인
print(interest_df)
```
