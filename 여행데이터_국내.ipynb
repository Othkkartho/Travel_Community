```python
import pandas as pd

# CSV 파일 리스트
csv_files = ["travel_20230814.csv", "travel_20230807.csv", "travel_20230821.csv", "travel_20230828.csv"]

# 각 CSV 파일을 읽어서 DataFrame 리스트 생성
df_list = [pd.read_csv(file) for file in csv_files]

# 모든 DataFrame 합치기
df = pd.concat(df_list, ignore_index=True)
df
```

```python
data2=[]

for i in df['AGRDE_FLAG_NM']:
  data2.append(i)

data2 = list(set(data2))
data2 = sorted(data2)

data2
# data2 = 연령별
# data1 = 전체데이터
# 도시 이름 리스트
korean_cities = ['서울', '경기', '부산', '충남', '충북',
                 '경남', '경북', '전남', '전북', '강원',
                 '제주도']
city = ['SU_TOUR_INTRST_VALUE',
       'GG_IC_TOUR_INTRST_VALUE', 'BS_TOUR_INTRST_VALUE',
       'CHN_DJ_TOUR_INTRST_VALUE', 'CB_TOUR_INTRST_VALUE',
       'GN_US_TOUR_INTRST_VALUE', 'GB_DG_TOUR_INTRST_VALUE',
       'JN_GJ_TOUR_INTRST_VALUE', 'JNBK_TOUR_INTRST_VALUE',
       'GW_TOUR_INTRST_VALUE', 'JJ_TOUR_INTRST_VALUE']
result = {}

for age in data2:
  # 연령대별로 데이터 필터링
  data_age = df[df['AGRDE_FLAG_NM'] == age]

  result[age] = {} # 연령대별 결과를 저장할 딕셔너리 초기화

  for j in range(len(city)):
    # 각 도시에 대한 관심도 계산 후 딕셔너리에 저장
    data1[city[j]] = pd.to_numeric(data1[city[j]], errors='coerce').astype(int)
    result[age][korean_cities[j]] = data1[city[j]].mean()

print(result)



```

```python
import matplotlib.pyplot as plt

# 결과 딕셔너리에서 연령대 가져오기
ages = list(result.keys())

plt.rcParams['font.family'] = 'Malgun Gothic'
# 각 연령대에 대해 막대 그래프 그리기
for age in ages:
    # 해당 연령대의 도시별 관심도 가져오기
    interests = result[age]

    # 도시 이름과 관심도로부터 리스트 생성
    cities = list(interests.keys())
    values = list(interests.values())

    plt.figure(figsize=(10, 5))  # 크기 설정

    plt.bar(cities, values)  # 막대 그래프 생성

    plt.title(f'{age}의 도시별 관심도')  # 제목 설정
    plt.xlabel('도시')  # x축 레이블 설정
    plt.ylabel('관심도')  # y축 레이블 설정

    plt.show()  # 그래프 출력
```
