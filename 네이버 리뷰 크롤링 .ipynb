```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import pandas as pd

# Chrome ë“œë¼ì´ë²„ ì´ˆê¸°í™”
driver = webdriver.Chrome()

# URL ì—´ê¸° (ì•žì„  ë‹µë³€ì—ì„œ ì–»ì€ urls ë°°ì—´ì„ ì‚¬ìš©)
korean_cities = ['ëŒ€í•œë¯¼êµ­', 'ê²½ë¶', 'ê²½ë‚¨', 'ì¶©ë‚¨', 'ì¶©ë¶', 'ê²½ê¸°',
                 'ì „ë‚¨', 'ì „ë¶', 'ì„œìš¸', 'ë¶€ì‚°', 'ê°•ì›',
                 'ì œì£¼']

data = {'ì§€ì—­': [], "ì œëª©": [], "ë‚´ìš©": []}  # ë°ì´í„° í”„ë ˆìž„ì„ ìœ„í•œ ë”•ì…”ë„ˆë¦¬ ì´ˆê¸°í™”

for idx, url in enumerate(all_urls):
    driver.get(url)

    try:
        # span íƒœê·¸ì˜ '_2tObC' í´ëž˜ìŠ¤ë¥¼ ê°€ì§„ ëª¨ë“  ìš”ì†Œ ëŒ€ê¸° ë° ì„ íƒ 
        elements_zPfVt = WebDriverWait(driver, 20).until(
            EC.presence_of_all_elements_located((By.CLASS_NAME, "zPfVt"))
        )

        # span íƒœê·¸ì˜ '_3XamX' í´ëž˜ìŠ¤ë¥¼ ê°€ì§„ ëª¨ë“  ìš”ì†Œ ëŒ€ê¸° ë° ì„ íƒ 
        elements_Fc1rA = WebDriverWait(driver, 10).until(
            EC.presence_of_all_elements_located((By.CLASS_NAME, "Fc1rA"))
        )
    except:
        driver.refresh()  # íŽ˜ì´ì§€ ìƒˆë¡œê³ ì¹¨

        try:
            elements_zPfVt = WebDriverWait(driver, 20).until(
                EC.presence_of_all_elements_located((By.CLASS_NAME, "zPfVt"))
            )
            elements_Fc1rA = WebDriverWait(driver, 10).until(
                EC.presence_of_all_elements_located((By.CLASS_NAME, "Fc1rA"))
            )
        except:
            elements_zPfVt = ['-']
            elements_Fc1rA = ['-']

    city_name_index = idx // 6   # í˜„ìž¬ ì¸ë±ìŠ¤ì— í•´ë‹¹í•˜ëŠ” ë„ì‹œëª… ì¸ë±ìŠ¤ ê³„ì‚° (0~59ê¹Œì§€ëŠ” korean_cities[0], 60~119ê¹Œì§€ëŠ” korean_cities[1] ë“±)
    city_name_index %= len(korean_cities)   # ë§Œì•½ ì¸ë±ìŠ¤ê°€ ë„ì‹œ ê°œìˆ˜ë³´ë‹¤ í¬ë©´ ì²˜ìŒë¶€í„° ë‹¤ì‹œ ì‹œìž‘
    
    city_name = korean_cities[city_name_index]
    
    title_text = elements_Fc1rA[0].text if type(elements_Fc1rA[0]) != str else ""  

    for element in elements_zPfVt:
        data['ì§€ì—­'].append(city_name)   # ì§€ì—­ëª… ì¶”ê°€
        data['ì œëª©'].append(title_text)   # ì œëª© ì¶”ê°€
        data['ë‚´ìš©'].append(element.text if type(element) != str else element)  # ë‚´ìš© ì¶”ê°€

df = pd.DataFrame(data)  # ë°ì´í„° í”„ë ˆìž„ ìƒì„±

print(df)  # ë°ì´í„° í”„ë ˆìž„ ì¶œë ¥

driver.quit()  # ë¸Œë¼ìš°ì € ì¢…ë£Œ

# DataFrame to csv file and xlsx file.
df.to_csv('naverRv.csv')
df.to_excel('naverRv.xlsx')
```

```python
import pandas as pd
from konlpy.tag import Okt
from collections import Counter
from wordcloud import WordCloud
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np

# URL ì—´ê¸° (ì•žì„  ë‹µë³€ì—ì„œ ì–»ì€ urls ë°°ì—´ì„ ì‚¬ìš©)
korean_cities = ['ëŒ€í•œë¯¼êµ­', 'ê²½ë¶', 'ê²½ë‚¨', 'ì¶©ë‚¨', 'ì¶©ë¶', 'ê²½ê¸°',
                 'ì „ë‚¨', 'ì „ë¶', 'ì„œìš¸', 'ë¶€ì‚°', 'ê°•ì›',
                 'ì œì£¼']

okt = Okt()

# ì›Œë“œí´ë¼ìš°ë“œ ëª¨ì–‘ ì´ë¯¸ì§€ ë¡œë“œ ë° ë³€í™˜ 
mask_image = np.array(Image.open('kor.png'))
plt.rcParams["font.family"] = "Malgun Gothic"

for city in korean_cities:
    # í•´ë‹¹ ë„ì‹œì˜ CSV íŒŒì¼ ë¡œë“œ 
    df_city = pd.read_csv(f"naver_{city}.csv")

    # í˜•íƒœì†Œ ë¶„ì„ ë° ë¶ˆìš©ì–´ ì œê±° 
    morphs = []
    for sentence in df_city['ë‚´ìš©']:
        morphs.extend(okt.nouns(sentence))

    count_morphs = Counter(morphs)

     # ì—¬ê¸°ì„œ ë¶ˆìš©ì–´ ë¦¬ìŠ¤íŠ¸ëŠ” ì‚¬ìš©ìžê°€ ì§ì ‘ ì •ì˜í•´ì•¼ í•©ë‹ˆë‹¤.
    stop_words = set(['ì˜','ê°€','ì´','ì€','ë“¤','ëŠ”','ì¢€','ìž˜','ê±','ê³¼',
             'ë„','ë¥¼','ìœ¼ë¡œ','ìž','ê³³','ìˆ˜','ê²ƒ','ì •ë§','ì•„ì´','ì‚¬ëžŒ','ì‹œê°„','ì¶”ì²œ'
                     ])
    
    processed_morphs_dict= {word: count for word, count in count_morphs.items() if word not in stop_words and len(word) > 1}

     # ìƒìœ„ 30ê°œ ë‹¨ì–´ë§Œ ì„ íƒí•˜ì—¬ ìƒˆë¡œìš´ ì‚¬ì „ ìƒì„± 
    processed_morphs_dict_top30=dict(sorted(processed_morphs_dict.items(), key=lambda item: item[1], reverse=True)[:30])
    
     # ì›Œë“œ í´ë¼ìš°ë“œ ìƒì„± 
    wc=WordCloud(font_path='c:/Windows/Fonts/malgun.ttf',
                  background_color="white", mask=mask_image)
     
    cloud=wc.generate_from_frequencies(processed_morphs_dict_top30)

    fig, ax=plt.subplots(1,1, figsize=(10,10))

    # ì›Œë“œí´ë¼ìš°ë“œ ê·¸ë¦¬ê¸°
    ax.imshow(cloud, interpolation='bilinear')

    # ë°°ê²½ ì´ë¯¸ì§€ ê·¸ë¦¬ê¸°
    ax.imshow(mask_image, cmap=plt.cm.Greens, interpolation='bilinear', alpha=0.15)

    ax.axis('off')
    plt.title(city)
    plt.show()
```
