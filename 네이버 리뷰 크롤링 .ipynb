```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

# Chrome 드라이버 초기화
driver = webdriver.Chrome()

# cityUrl 배열 초기화 (예시)

all_urls = []  # 모든 결과를 저장할 배열 초기화

for url in cityUrls:
    driver.get(url)

    # li 태그의 'S0Ns3 DyAcu' 클래스를 가진 모든 요소 대기 및 선택 
    elements = WebDriverWait(driver, 10).until(
        EC.presence_of_all_elements_located((By.CSS_SELECTOR, "li.S0Ns3.DyAcu"))
    )

    urls = []  # 해당 도시의 결과를 저장할 배열 초기화

    for element in elements:
        data_tri_doc_id = element.get_attribute('data-tri-doc-id')  # 각 요소의 data-tri-doc-id 값 추출
        
        new_url = f"https://pcmap.place.naver.com/place/{data_tri_doc_id}/review/visitor"  # 새 URL 생성
        urls.append(new_url)  # 해당 도시의 배열에 추가

    all_urls.extend(urls)  # 전체 결과에 해당 도시의 결과 추가

print(all_urls)  # 최종 결과 출력

driver.quit()
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import pandas as pd

# Chrome 드라이버 초기화
driver = webdriver.Chrome()

# URL 열기 (앞선 답변에서 얻은 urls 배열을 사용)

data = {'제목': [], '내용': []}  # 데이터 프레임을 위한 딕셔너리 초기화

for url in urls:
    driver.get(url)

    # span 태그의 'zPfVt' 클래스를 가진 모든 요소 대기 및 선택 
    elements_zPfVt = WebDriverWait(driver, 10).until(
        EC.presence_of_all_elements_located((By.CLASS_NAME, "zPfVt"))
    )

    # span 태그의 'Fc1rA' 클래스를 가진 모든 요소 대기 및 선택 
    elements_Fc1rA = WebDriverWait(driver, 10).until(
        EC.presence_of_all_elements_located((By.CLASS_NAME, "Fc1rA"))
    )

    title_text = elements_Fc1rA[0].text if elements_Fc1rA else ""  # 제목 추출 (없으면 빈 문자열)

    for element in elements_zPfVt:
        data['제목'].append(title_text)   # 제목 추가
        data['내용'].append(element.text)  # 내용 추가

df = pd.DataFrame(data)  # 데이터 프레임 생성

print(df)  # 데이터 프레임 출력

driver.quit()  # 브라우저 종료

# 필요하다면 df.to_csv('filename.csv') 등으로 파일에 저장할 수 있습니다.
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import pandas as pd

# Chrome 드라이버 초기화
driver = webdriver.Chrome()

# URL 열기 (앞선 답변에서 얻은 urls 배열을 사용)
korean_cities = ['대한민국', '경북', '경남', '충남', '충북', '경기',
                 '전남', '전북', '서울', '부산', '강원',
                 '제주']

data = {'지역': [], "제목": [], "내용": []}  # 데이터 프레임을 위한 딕셔너리 초기화

for idx, url in enumerate(all_urls):
    driver.get(url)

    try:
        # span 태그의 '_2tObC' 클래스를 가진 모든 요소 대기 및 선택 
        elements_zPfVt = WebDriverWait(driver, 20).until(
            EC.presence_of_all_elements_located((By.CLASS_NAME, "zPfVt"))
        )

        # span 태그의 '_3XamX' 클래스를 가진 모든 요소 대기 및 선택 
        elements_Fc1rA = WebDriverWait(driver, 10).until(
            EC.presence_of_all_elements_located((By.CLASS_NAME, "Fc1rA"))
        )
    except:
        driver.refresh()  # 페이지 새로고침

        try:
            elements_zPfVt = WebDriverWait(driver, 20).until(
                EC.presence_of_all_elements_located((By.CLASS_NAME, "zPfVt"))
            )
            elements_Fc1rA = WebDriverWait(driver, 10).until(
                EC.presence_of_all_elements_located((By.CLASS_NAME, "Fc1rA"))
            )
        except:
            elements_zPfVt = ['-']
            elements_Fc1rA = ['-']

    city_name_index = idx // 6   # 현재 인덱스에 해당하는 도시명 인덱스 계산 (0~59까지는 korean_cities[0], 60~119까지는 korean_cities[1] 등)
    city_name_index %= len(korean_cities)   # 만약 인덱스가 도시 개수보다 크면 처음부터 다시 시작
    
    city_name = korean_cities[city_name_index]
    
    title_text = elements_Fc1rA[0].text if type(elements_Fc1rA[0]) != str else ""  

    for element in elements_zPfVt:
        data['지역'].append(city_name)   # 지역명 추가
        data['제목'].append(title_text)   # 제목 추가
        data['내용'].append(element.text if type(element) != str else element)  # 내용 추가

df = pd.DataFrame(data)  # 데이터 프레임 생성

print(df)  # 데이터 프레임 출력

driver.quit()  # 브라우저 종료

# DataFrame to csv file and xlsx file.
df.to_csv('naverRv.csv')
df.to_excel('naverRv.xlsx')
```

```python
import pandas as pd
from konlpy.tag import Okt
from collections import Counter
from wordcloud import WordCloud
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np

# URL 열기 (앞선 답변에서 얻은 urls 배열을 사용)
korean_cities = ['대한민국', '경북', '경남', '충남', '충북', '경기',
                 '전남', '전북', '서울', '부산', '강원',
                 '제주']

okt = Okt()

# 워드클라우드 모양 이미지 로드 및 변환 
mask_image = np.array(Image.open('kor.png'))
plt.rcParams["font.family"] = "Malgun Gothic"

for city in korean_cities:
    # 해당 도시의 CSV 파일 로드 
    df_city = pd.read_csv(f"naver_{city}.csv")

    # 형태소 분석 및 불용어 제거 
    morphs = []
    for sentence in df_city['내용']:
        morphs.extend(okt.nouns(sentence))

    count_morphs = Counter(morphs)

     # 여기서 불용어 리스트는 사용자가 직접 정의해야 합니다.
    stop_words = set(['의','가','이','은','들','는','좀','잘','걍','과',
             '도','를','으로','자','곳','수','것','정말','아이','사람','시간','추천'
                     ])
    
    processed_morphs_dict= {word: count for word, count in count_morphs.items() if word not in stop_words and len(word) > 1}

     # 상위 30개 단어만 선택하여 새로운 사전 생성 
    processed_morphs_dict_top30=dict(sorted(processed_morphs_dict.items(), key=lambda item: item[1], reverse=True)[:30])
    
     # 워드 클라우드 생성 
    wc=WordCloud(font_path='c:/Windows/Fonts/malgun.ttf',
                  background_color="white", mask=mask_image)
     
    cloud=wc.generate_from_frequencies(processed_morphs_dict_top30)

    fig, ax=plt.subplots(1,1, figsize=(10,10))

    # 워드클라우드 그리기
    ax.imshow(cloud, interpolation='bilinear')

    # 배경 이미지 그리기
    ax.imshow(mask_image, cmap=plt.cm.Greens, interpolation='bilinear', alpha=0.15)

    ax.axis('off')
    plt.title(city)
    plt.show()
```
